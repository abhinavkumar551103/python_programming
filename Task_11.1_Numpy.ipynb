# OTP Generator

```python
import random
def otp_generator(Length = 4):
    """Generate a numeric OTP of a specified length."""
    digits = "0123456789"
    otp = ''.join(random.choice(digits) for _ in range(Length))
    return otp
# Example usage:
otp_length = 6
otp = otp_generator(otp_length)
print(f'Your OTP is: {otp}')
```

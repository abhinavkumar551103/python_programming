# Tuple ( )

```python
tup1 = () # Emplty tuple

# Properties Of tuple

# Immutable
# Ordered (elements remain in the order they were added)
# Can be used as Dictionary keys 
# Slicing Support
# Indexing[] Support
# Supports Duplicate values and Heterogenous Data (Mixed Data Types)
# Can store nested objects : tuple,List,Sets also
# Built-in Methods are: count() ,index()
```

```python
tup_int = (10,20,30,50,60)
tup_float = (10.5,24.36,124.1,32.85,28.36)
tup_String = ('one','Two','Three','Five','six')
tup_Mixed = (201010, 'Anand','gupta',36.56)
tup_Nested = ('Asif', 50,(75,85,95,105),(90,103))
tup_Multiple = ('Asif',25,[50,100],{'John','David'},(99,103,128))
```

```python
print(len(tup_Multiple))
```

```python
tup_int[0] # first (0) index of integer tuple
```

```python
tup_float[0] # First (0) index of float tuple
```

```python
tup_String[0] # first (0) index of the string tuple
```

```python
tup_Nested[2][1] # Get value from nested tuple value
```

```python
tup_String[-1] # last value of the String tuple
```

```python
print(tup_Mixed[1])
print(tup_Mixed[1][0]) # Access the variable of the tuple element
```

# Loop through a Tuple

```python
mytuple = ('ONE_1','TWO_2','THREE_3','FOUR_4','FIVE_5','SIX_6')
```

```python
for i in mytuple:
    print(i)
```

```python
for i in enumerate(mytuple):
    print(i)
```

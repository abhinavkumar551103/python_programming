# Copy( )

```python
myset = {'One','Two','Three','Four','Five','Six','Seven','Eight'}
myset
```

```python
myset_1 = myset # create a new reference my set
myset_1 
```

```python
id(myset), id(myset_1) # the location address of myset and myset_1 is same
```

```python
myset_Copy = myset.copy()
myset_Copy  # Create a copy of the set
```

```python
id(myset_Copy), id(myset) # Both sets have different memory addresses because copy() creates a new set object.
```

```python
myset_Copy.add('Nine')
myset_Copy # Only the myset_Copy is added 'nine' the myset does not impacted
```

```python
myset # Does not get impacted
```

# Set Operations

```python
# UNION
# INTERSECTION
# DISJOINT
# DIFFERENCE
# SYMMETRIC DIFFERENCE
# SUBSET
# SUPERSET
```

```python
# UNION OPERATION
```

```python
NumSet_1 = {1,2,3,4,5,6}
NumSet_2 = {4,5,6,7,8,9,10}
NumSet_3 = {8,9,10,11,12,13,14,15}
```

```python
NumSet_1.union(NumSet_2) # Its Join all elements of set A and set B
```

```python
NumSet_1.union(NumSet_2, NumSet_3) # We can apply on multiple set
```

```python
NumSet_1 | NumSet_2 # Also denoted by {|}  
```

```python
# INTERSECTION OPERATION
```

```python
NumSet_1.intersection(NumSet_2) # Intersection of A and B (Common items in both sets) 
```

```python
NumSet_2.intersection(NumSet_3)
```

```python
NumSet_2 & NumSet_3 # Also denoted by {&} 
```

```python
# DIFFERENCE OPERATION
```

```python
print(NumSet_1)
print(NumSet_2)
print(NumSet_3)
```

```python
NumSet_1.difference(NumSet_2) # set of elements that are only in A but not in B
```

```python
NumSet_2.difference(NumSet_3)
```

```python
NumSet_1 - NumSet_2 # Also denoted by {-}
```

```python
# SYMMETRIC DIFFERENCE OPERATION
```

```python
NumSet_1.symmetric_difference(NumSet_2)  # Symmetric difference (Set of elements in A and B but not in both.
```

```python
NumSet_2.symmetric_difference(NumSet_3)
```

```python
NumSet_1 ^ NumSet_2 # Also denoted by {^}
```

# Loop through a dictionary

```python
mydict_profile = {'Name': 'Harsh', 'ID': 125696, 'DOB': 1991, 'Address': 'Chennai'}
mydict_profile
```

```python
for i in mydict_profile:
    print(i , ':' , mydict_profile[i]) # keys value pair
```

```python
for i in enumerate(mydict_profile):
    print(i) # all keys
```

```python
# All / Any 
# The all() method returns: 
# True - If all all keys of the dictionary are true 
# False - If any key of the dictionary is false 
# The any() function returns True if any key of the dictionary is True. If not, any() returns False
```

```python
mydict_profile
```

```python
any(mydict_profile) # return false is all values are false neither one statement is true
```

```python
all(mydict_profile) # Will Return false as one value is false (Value 0) 
```

# All Build-in Functions of List

```python
shopping_list = ["Milk", "Bread", "Eggs", "Rice", "Apples"]
```

```python
shopping_list.append("Oats")
shopping_list  # adding item using append() at the end of the list
```

```python
shopping_list.insert(2,"Butter")
shopping_list # add item using insert(idx,element) at particular index
```

```python
My_cart = [] # create empty list
```

```python
My_cart.extend(shopping_list)  # Add all shopping items to the cart
```

```python
My_cart
# shopping_list
```

```python
My_cart.sort()
My_cart  # sort theall items in list
```

```python
My_cart.sort(reverse=True)
My_cart # Basically reverse using sort so that its reverse the list after comparing the element
```

```python
My_cart.reverse()  # Reverse the current order of the list (no sorting)
My_cart
```

```python
My_cart.remove("Rice") # its remove the item from the list , but if its present more then once then it remove the first one next one are remains
```

```python
My_cart.pop() # basically its remove the item at particular index if given otherwise by default it consider the last element of the list and remove from the list
```

```python
My_cart.count("Milk") # Basically cont the element that how many times its prent in list.
```

```python
My_cart.index("Apples") # Basically its tells the index value on the element in the list
```

```python
Total_bill = My_cart.copy()
Total_bill # # Create a copy of My_cart in a new list
```

```python
My_cart.clear()
My_cart # Its remove all items from the list.
```

```python
len(Total_bill)        # Number of items
```

```python
del shopping_list[1]      # Delete using index
shopping_list
```

```python
Total_bill[0:3] # slicing
```
